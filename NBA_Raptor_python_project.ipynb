```python
#line  chart showing total_raptor of Stephen Curry across all seasons

# Filter data for Stephen Curry
curry_data = df_player[df_player['player_name'] == 'Stephen Curry']

# Select relevant columns
curry_raptor = curry_data[['season', 'raptor_total']]

# Sort data by season
curry_raptor = curry_raptor.sort_values(by ='season')


# Plot the line chart
plt.figure(figsize=(10, 6))
plt.plot(curry_raptor['season'], curry_raptor['raptor_total'], marker='o', linestyle='-', color='b')
plt.title("Stephen Curry's Total RAPTOR by Season")
plt.xlabel("Season")
plt.ylabel("Total RAPTOR")
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Create a scatter plot comparing offensive_raptor vs defensive_raptor

# Filter out rows with missing or zero values in relevant columns
df_filtered = df_player[(df_player['raptor_offense'] != 0) & (df_player['raptor_defense'] != 0)]



# Scatter plot comparing offensive_raptor vs defensive_raptor

plt.figure(figsize=(10, 6))
plt.scatter(df_filtered['raptor_offense'], df_filtered['raptor_defense'], alpha=0.5, color='b')
plt.title('NBA Players: Offensive vs. Defensive RAPTOR')
plt.xlabel('Offensive RAPTOR')
plt.ylabel('Defensive RAPTOR')
plt.grid(True)
plt.tight_layout()
plt.show()

```

```python
# Use a bar chart to show the top 5 players by `total_war` for season 2016 

# Filter data for the 2016 season
df_2016 = df_player[df_player['season'] == 2016]

# Sort by WAR and select top 5 players
top_5_war = df_2016[['player_name', 'war_total']].sort_values(by='war_total', ascending=False).head(5)

# plot a bar chart
plt.figure(figsize=(10, 6))
plt.barh(top_5_war['player_name'], top_5_war['war_total'], color='b')
plt.xlabel('Total WAR')
plt.title('Top 5 NBA Players by Total WAR in 2016 Season')
plt.gca().invert_yaxis()  # Invert y-axis to display the highest value at the top
plt.tight_layout()
plt.show()
```
